```python
import matplotlib.pyplot as plt
padding_accuracies = [.83, .82]
padding_titles = ["No Padding", "Padding"]

plt.bar(padding_titles, padding_accuracies)
plt.ylim(.75, .85)
plt.ylabel("Test Accuracy")
plt.title("Padding")
```

```python
stride_accuracies = [.83, .52]
stride_titles = ["1", "2"]

plt.bar(stride_titles, stride_accuracies)
plt.ylim(.5, .9)
plt.title("Stride")
plt.ylabel("Test Accuracy")
```

```python
stride_accuracies = [.82, .80, .75, .57, 0.4]
stride_titles = list(range(1, 6))

plt.plot(stride_titles, stride_accuracies)
# plt.ylim(.5, .9)
plt.title("Dilation")
plt.ylabel("Test Accuracy")
plt.xlabel("Dilation Rate")
plt.xticks(stride_titles)
```
